Булево индексирование массивов позволяет выбирать элементы массива по логической маске. Часто это используется для выбора или изменения элементов, удовлетворяющих некоторому условию:

```python
a = np.array([[1,2], [3, 4], [5, 6]])

# Создадим и выведем логическую маску
bool_idx = (a > 3)  

print(bool_idx)
print(bool_idx.dtype)
print(type(bool_idx))
```

```python
# Выведем результат отбора
print(a[bool_idx])

# Разумеется, можно использовать inplace:
a[a > 2] = 2
print(a)
```

Часто полезно получать *индексы* элементов массива, удовлетворяющих некоторым условиям. Есть несколько методов для этого
-   [`argmax`](https://numpy.org/doc/stable/reference/generated/numpy.argmax.html) (получить индекс максимального элемента массива)
-   [`argmin`](https://numpy.org/doc/stable/reference/generated/numpy.argmin.html) (получить индекс минимального элемента массива)
-   [`argsort`](https://numpy.org/doc/stable/reference/generated/numpy.argsort.html) (получить набор индексов, сортирующий массив по возрастанию)
-   [`where`](https://numpy.org/doc/stable/reference/generated/numpy.where.html) (получить индексы элементов, удовлетворяющих некоторому условию)

```python
a = np.array([1, 8, 9, -3, 2, 4, 7, 9])

print(np.argmax(a))

print(np.argmin(a))

print(np.argsort(a))

# Отсортируем индексы в обратном порядке
print(np.argsort(a)[::-1])

# Получим индексы элементов, удовлетворяющих условию.
# Функция также возвращает сами элементы, поэтому выберем только срез [0]
print(np.where(a > 5)[0])

# Получим индексы всех элементов массива, равных максимальному
print(np.where(a >= a[np.argmax(a)])[0])
```

### Арифметические типы

Как уже упоминалось, каждый `numpy.array` хранит однотипные элементы. `numpy` предоставляет большое количество численных типов данных. Обычно тип данных успешно определяется автоматически при инициализации массива, но можно указать его явно:

```python
x = np.array([1, 2])  # Автоматическое определение
y = np.array([1.0, 2.0])  # Автоматическое определение
z = np.array([1, 2], dtype=np.float16)  # Задаём тип явно
w = np.array([0, 1], dtype=np.bool) # Задаём тип явно

print(x.dtype, y.dtype, z.dtype, w.dtype)
```

В этом семестре мы преимущественно будем работать с `fp32` арифметикой и потому на время забудем про числовой тип.

Однако, как мы увидим в следующем семестре, использование сокращённых типов - `fp16`, `fp8` и даже `fp4` полезно и очень распространено в современных моделях. 

Это связано с тем, что большим сетям совершенно не нужна точность `fp32` и тем более `fp64` для работы - важнее архитектура, количество параметров и ресурсы, затраченные на обучение и инференс. Так, например, создатели DeepSeek добились схожего с openai-o1 качества при намного меньшей ресурсоёмкости и меньшей стоимости обучения в том числе за счёт широкого использования `fp16` и `fp8` арифметики.

С разрядностью арифметики связан ещё один нюанс - с распространением сокращённой арифметики поставщики оборудования зачастую стали указывать производительность (FLOPS) в сокращённых типах, а не в `fp64` - что, как вы понимаете, даёт очень существенный прирост производительности) Так, например Nvidia PROJECT Digits обещает целый PetaFLOPS ($10^{15}$) в компактном устройстве - но это PFLOPS в `fp4`, что вполне сравнимо не только с графическими ускорителями, но и с процессорами.

### Broadcasting

Broadcasting - это механизм, который позволяет `numpy` автоматически приводить массивы разных форм при выполнении математических операций.

Например, в линейной алгебре мы можем складывать только матрицы с одинаковыми размерностями. Однако если `numpy` выполняет сложение матриц с разными размерностями, он попытается привести их размерности к одной, так, что операция будет выполнена без ошибки:

![](https://sebastianraschka.com/images/blog/2020/numpy-intro/broadcasting-1.png)

![](https://sebastianraschka.com/images/blog/2020/numpy-intro/broadcasting-2.png)

Broadcasting выполняется по следующим правилам:

**Правило 1**: Если два массива отличаются по количеству размерностей, `shape` массива с меньшей размерностью будет дополняться единицами, начиная с последней размерности

```python
a = np.array([1,2,3])         # имеет shape (3,)
b = np.array([[4], [5], [6]]) # имеет shape (3,1)
c = a + b                     # получает shape (3,3) 
print(c.shape)
```

**Правило 2**: Если форма двух массивов отличается по какой-то размерности и один массив имеет длину 1 по этой размерности, то этот массив по этой размерности при выполнении операций будет дополняться своими копиями:

```python
a = np.array([[1],[2],[3]])         # shape (3,1)
b = np.array([[4,5], [6,7], [8,9]]) # shape (3,2)
c = a + b                           # shape (3,2)

print(c)
```

**Правило 3**: Если какая-то размерность не совпадает и не выполняется ни одно из предыдущих правил, вызывается ошибка:

```python
a = np.array([[1],[2],[3],[4]])      # shape (4,1)
b = np.array([[4,5], [6,7], [8,9]])  # shape (3,2)
c = a + b                            # ValueError: operands could not be broadcast
```

Более подробно процедура описана в [документации](http://docs.scipy.org/doc/numpy/user/basics.broadcasting.html).

Функции, поддерживающие приведение размерностей, называются *универсальными*. Рассмотрим ещё несколько примеров таких функций:

Functions that support broadcasting are known as universal functions. You can find the list of all universal functions in the [documentation](http://docs.scipy.org/doc/numpy/reference/ufuncs.html#available-ufuncs).



![](http://jalammar.github.io/images/numpy/numpy-array-broadcast.png)


![](http://jalammar.github.io/images/numpy/numpy-matrix-broadcast.png)

И ещё несколько примеров

```python
# Вычислим внешнее произведение векторов
v = np.array([1,2,3])  # v shape (3,)
w = np.array([4,5])    # w shape (2,)

# Чтобы вычислить внешнее произведение, нужно сначала преобразовать v к столбцу
# с shape (3, 1); после чего при умножении он автоматически будет приведён привести его к w
# по последней размерности, в результате получится внешнее произведение векторов с shape (3, 2):

print(np.reshape(v, (3, 1)) * w)
```

```python
# Добавим вектор к каждой строке матрицы
x = np.array([[1,2,3], [4,5,6]])

print(x + v)
```

```python
# Умножим матрицу на константу:
# x имеет shape (2, 3). Numpy считает скаляры массивами с shape ();
# который может быть приведён к shape (2, 3):
print(x * 2)
```

Broadcasting часто делает код более быстрым (время выполнения внешней самописной процедуры приведения размерностей будет больше, чем время автоматического приведения) и более читаемым, однако в то же время может приводить к множеству ошибок при невнимательном использовании.

За более подробной справкой обращайтесь к [numpy reference](http://docs.scipy.org/doc/numpy/reference/).

Её функционал очень широк, мы рассмотрим только малую его часть, необходимую для понимания синтаксиса работы с библиотекой.

Для остальных задач рекомендую обращаться [документации](http://matplotlib.org/stable) и [примерам](https://matplotlib.org/stable/gallery/index). Последние особенно полезны, поскольку на практике 90% графиков в `matplotlib` строятся адаптацией этих примеров под свои нужды.

```python
# Стандартный способ импорта основного модуля matplotlib
import matplotlib.pyplot as plt
```

Метод `plt.plot` позволяет построить линейный график:

```python
# Создадим массив точек на оси X и вычислим соответствующие значения исследуемой функции:
x = np.arange(0, 3 * np.pi, 0.6
              )
y = np.sin(x)

# Построим и выведем график
plt.plot(x, y)

plt.show()
```

На один график можно нанести множество линий:

```python
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, label='Sin')
plt.plot(x, y_cos, label='Cos')
plt.legend() # Отобразить метки линий, указанные выше

plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sin and Cos')

plt.show()
```

Другой базовый метод - `plt.scatter`

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.randint(0, 2, 50)

plt.scatter(x, y, c=colors)

plt.show()
```

Часто бывает нужно собрать несколько графиков на одной фигуре - для этого используют модуль `plt.subplot`:

```python
# Снова 
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Создадим холст c заданным размером
fig = plt.figure(figsize=(12, 4))
# аргумент figsize принимает кортеж - ширину и высоту в дюймах

# Создадим решётку графиков высоты 1 и ширины 2,
# выбрав первый график в качестве активного:
plt.subplot(1, 2, 1)

# Отрисуем первый график
plt.plot(x, y_sin)
plt.title('Sin')

# Выберем и отрисуем второй график
plt.subplot(1, 2, 2)
plt.plot(x, y_cos)
plt.title('Cos')

# Выведем график
plt.show()
```

Функция `imshow` пригодится для отображения любого двумерного массива в виде картинки:

```python
noise = np.random.random(size=(128, 128))

plt.imshow(noise)
plt.show()
```
