# Автоматическое дифференцирование в pytorch

*НЛО*:
В тензорном исчислении, тензор можно представить следующей формулой:

    тензор = линейный оператор + закон его преобразования при замене координат

Тензоры в `pytorch` не являются тензорами в этом смысле слова - прежде всего, не идёт речи о каких-либо преобразованиях координат.

Но аналогия (и названия) вполне уместны, потому что "формула" `torch.tensor` очень похожа

    `torch.tensor` = линейный оператор + закон его автоматического дифференцирования

Поскольку в тензорном анализе преобразование тензоров при замене координат тоже происходит с участием производных, вероятно, создатели `pytorch` при выборе названия для этого класса, вероятно, руководствовались похожей логикой.
